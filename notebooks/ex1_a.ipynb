```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = Image.open('..\\resources\\lena.tif')
```

Se crea una funci칩n que para cada pixel, devuelve su valor en negativo, teniendo como par치metro el m치ximo valor de intensidad de cada pixel (por defecto 256)

```python
def negative_transf(r:float, L=256) -> float:
    s = L - 1 - r
    return s
```

Se obtiene el negativo de la imagen original, aplicando la transformaci칩n elemento a elemento

```python
im_data = list(im.getdata())
neg_im = np.array(list(map(negative_transf, im_data))).reshape(im.size)
```

Se grafican lado a lado la imagen original y su negativo

```python
fig = plt.figure()
fig.set_size_inches(30, 15)
rows = 1
cols = 2

fig.add_subplot(rows, cols, 1)
plt.axis('off')
plt.imshow(np.array(list(im.getdata())).reshape(im.size), cmap='gray')
plt.title('Original')

fig.add_subplot(rows, cols, 2)
plt.axis('off')
plt.imshow(neg_im, cmap='gray')
plt.title('Negative')

plt.show()
```
