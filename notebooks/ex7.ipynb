# Ecualización de un histograma
La ecualización de un histograma se utiliza para lograr un mejor aprovechamiento del rango dinámico de una imagen, y así corregir errores en la toma de la foto, como una mala exposición (brillo no balanceado), o para mejorar el contraste de la misma.

A continuación se muestran algunas imágenes antes y despues de una ecualización, con sus respectivos histogramas en la parte superior izquierda. También, superpuesto a las imagenes ecualizadas, se presenta la transformación utilizada. Las mismas fueron logradas a través de la aplicación Lightroom para edición de imágenes.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
rocks_before = mpimg.imread('..\\resources\\rocks-before.jpeg')
rocks_after = mpimg.imread('..\\resources\\rocks-after.jpeg')
scala_before = mpimg.imread('..\\resources\\scala-before.jpeg')
scala_after = mpimg.imread('..\\resources\\scala-after.jpeg')
window_before = mpimg.imread('..\\resources\\window-before.jpeg')
window_after = mpimg.imread('..\\resources\\window-after.jpeg')
```

```python
fig = plt.figure()
fig.set_size_inches(30, 15)
rows = 1
cols = 2

fig.add_subplot(rows, cols, 1)
plt.axis('off')
plt.imshow(rocks_before, cmap='gray')
plt.title('Original')

fig.add_subplot(rows, cols, 2)
plt.axis('off')
plt.imshow(rocks_after, cmap='gray')
plt.title('Equalised')

plt.show()
```
